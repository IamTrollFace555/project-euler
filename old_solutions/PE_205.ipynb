```python
arr_1 = [1, 1, 1, 1]
for _ in range(8):
    arr_1 = np.convolve(arr_1, [1, 1, 1, 1])
arr_1 = np.array([arr_1])/sum(arr_1)

arr_2 = [1, 1, 1, 1, 1, 1]
for _ in range(5):
    arr_2 = np.convolve(arr_2, [1, 1, 1, 1, 1, 1])
arr_2 = np.array([arr_2])/sum(arr_2)

print(round(np.sum(np.triu(arr_1.T @ arr_2, k=1)), 7))
```
