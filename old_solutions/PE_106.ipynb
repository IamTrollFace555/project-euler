```python
from itertools import chain, combinations
```

```python
def powerset(iterable):
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1)))

def get_pairs(s):
    subsets = powerset(s)
    setS = set(s)
    pairs = set()

    for idx, B in enumerate(subsets):
        for C in powerset(setS.difference(set(B))):
            if (C, B) not in pairs:
                pairs.add((B, C))
    return pairs
```

```python
# s = list(range(12))
s = [0, 1, 2, 3]
pairs = get_pairs(s)

# print(len(pairs))
for pair in pairs:
    print(pair)
```
