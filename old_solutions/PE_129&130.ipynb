```python
%%time

limit = 10**6

def solve():
    n = limit + 1
    repunit = 1
    count = 1
    nums = set((1, 7, 9))

    while True:
        while repunit != 0:
            repunit = (10 * repunit + 1) % n
            count += 1

        if count > limit:
            return n

        if n%10 in nums:
            n += 2

        elif n%10 == 3:
            n += 4

        repunit = 1
        count = 1

solve()
```

```python
def A(n):
    repunit = 1
    count = 1
    while repunit != 0:
        repunit = (10 * repunit + 1) % n
        count += 1

    return count


def is_prime(number):
    if number <= 1 or number%2 == 0:
        return False

    for i in range(3, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True
```

```python
def factores_primos(n:int) -> list:
    facts = set()                  # Inicializamos una lista para almacenar los factores primos
    p = 2                       # Empezamos verificando desde 2
    while n != 1:
        if n % p == 0:
            facts.add(p)
            while n % p == 0:       # si p|n, añadimos p a la lista de factores primos y repetimos hasta que p ∤ n.
                n = n//p
        p += 1
    return facts
```
