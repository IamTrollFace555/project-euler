```python
%%time
manhattan = 0
r = 10**4
r2 = r**2
count = 0
sqrt3r = 3**0.5 * r
for z in range(-r, r+1):
    upper = int((r2 - z**2)**0.5)

    for y in range(-upper , upper + 1):
        x = (r2 - z**2 - y**2) ** 0.5
        if x == 0:
            count += 1
            manhattan += abs(y) + abs(z)
        elif x % 1 == 0 :
            count += 2
            # print(f"x: {x}, y: {y}, z: {z}")
            manhattan += 2* (abs(x) + abs(y) + abs(z))

print(count)
print(manhattan)
```

```python
# Encontrar la descomposición prima de un número
def factores_primos(n:int) -> list:
    facts = {}                  # Inicializamos una lista para almacenar los factores primos
    p = 2                       # Empezamos verificando desde 2
    while n != 1:
        while n % p == 0:       # si p|n, añadimos p a la lista de factores primos y repetimos hasta que p ∤ n.
            n = n//p
            if p in facts.keys():
                facts[p] += 1
            else: facts[p] = 1
        p += 1
    return facts
```

```python
factores_primos(45)
```

```python
def count_pyth_triples(n):
    prime_dict = factores_primos(n)
    total = 1
    if 2 in prime_dict.keys():
        total *= (prime_dict[2] + 1)
        prime_dict[2] = 0

    for _, value in prime_dict.items():
        total *= (2*value + 1)

    return total

```

```python
count_pyth_triples(45)
```
