```python
max_length = 4
ways = [1] * (max_length + 1)
# ways[4] = [1, 1, 1, 1]

for length in range(3, max_length + 1):
    print("====================")
    print("Block Length: ", length)
    for i in range(length, max_length + 1):
        print("==========")
        print("Track Length: ", i)
        fits = i - length + 1 # Amount of ways the block can fit into the large space
        print("fits:", fits)
        limit = fits//2 if fits%2 == 0 else fits//2 + 1
        print("limit:", limit)
        for j in range(limit):
            print("=====")
            print("j: ", j)

            left = max(j - 1, 0)
            right = max(i - length - j - 1, 0)

            print("left: ", left)
            print("right: ", right)

            ways[i] += ways[left] * ways[right]

    print(ways)

print(ways)

print(ways[max_length])
```

```python
coins = [1, 2, 5, 10, 20, 50, 100, 200]
ways = [0] * 201
ways[0] = 1

for coin in coins:
    for i in range(coin, 201):
        ways[i] += ways[i-coin]

print(ways[200])
```
