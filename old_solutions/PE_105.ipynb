```python
from itertools import chain, combinations
```

```python
def process_set(s):
    elems = s.replace("\n", "").split(",")
    return list(map(int, elems))

with open("/content/0105_sets.txt") as file:
    lines = file.readlines()
    sets = list(map(process_set, lines))
```

```python
def powerset(iterable):
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(1, len(s)+1)))

def check_set(s):
    subsets = powerset(s)
    setS = set(s)
    # print("=================")

    for idx, B in enumerate(subsets):
        for C in powerset(setS.difference(set(B))):

            if sum(B) == sum(C):
                # print(B, sum(B))
                # print(C, sum(C))
                return False

            if len(B) > len(C) and sum(B) < sum(C):
                # print(B, sum(B))
                # print(C, sum(C))
                return False
    return True
```

```python
%%time

total = 0
for s in sets:
    if check_set(s):
        total += sum(s)

print(total)
```
