```python
from itertools import combinations
from functools import reduce
from operator import mul


nums = [*range(1, 16)]

combs = []
for i in range(2,8):
    combs += list(combinations(nums, i))

total = 0
for comb in combs:
    total += reduce(mul, comb)

fact16 = reduce(mul, [*range(2, 17)])
fact5 = 120

print(fact16/total)
```
