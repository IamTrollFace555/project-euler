```python
def Mobius(n):
    prime = [1] * (n + 1)
    mobius = [1] * (n + 1)

    for i in range(2, n + 1):
        if not prime[i]:
            continue
        mobius[i] = -1
        for j in range(2, n // i + 1):
            prime[i * j] = 0
            mobius[i * j] *= -1
        for j in range(1, n // (i * i) + 1):
            mobius[j * i * i] = 0
    return mobius
```

```python
def square_free(n):
    sq = int(n**0.5)
    mobius = Mobius(sq)

    s = 0
    for i in range(1, sq + 1):
        s += mobius[i] * (n // (i * i))
    return s
```

```python
%%time
print(square_free(2**50))
```
